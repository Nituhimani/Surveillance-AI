```python
from ultralytics import YOLO
import cv2
import json

# Load your trained model
model = YOLO("/content/best_activity.pt")

# Input video path
input_video = "/content/south_africa.mp4"
output_video = "/content/activity_output.mp4"
output_json = "/content/activity_output.json"

# OpenCV video reader and writer
cap = cv2.VideoCapture(input_video)
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
out = cv2.VideoWriter(output_video, fourcc, int(cap.get(5)),
                      (int(cap.get(3)), int(cap.get(4))))

frame_id = 0
results_list = []  # for JSON

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Run inference with the model
    results = model(frame, conf=0.5)  # set confidence threshold

    detections = []

    # Process results
    for r in results:
        for box in r.boxes:
            # Bounding box
            x1, y1, x2, y2 = box.xyxy[0].cpu().numpy()
            conf = float(box.conf[0])
            cls = int(box.cls[0])
            label = model.names[cls]

            # Save detection for JSON
            detections.append({
                "frame": frame_id,
                "label": label,
                "confidence": conf,
                "bbox": [float(x1), float(y1), float(x2), float(y2)]
            })

            # Draw box on frame
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)),
                          (0, 255, 0), 2)
            cv2.putText(frame, f"{label} {conf:.2f}", (int(x1), int(y1)-10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)

    results_list.extend(detections)

    # Write processed frame
    out.write(frame)
    frame_id += 1

cap.release()
out.release()

# Save detections to JSON
with open(output_json, "w") as f:
    json.dump(results_list, f, indent=4)

print(f"Processing complete! Output video: {output_video}, JSON: {output_json}")
```

## Define prediction function

### Subtask:
Create a Python function that takes the input video path, performs object detection using the `best_activity.pt` model, and saves the output video.


**Reasoning**:
Define a Python function to process the video, load the model, read the video, initialize the video writer, loop through frames, perform inference, draw bounding boxes, write frames, and release resources, then return the output video path.



```python
def process_video_with_yolo(input_video_path):
    """
    Processes a video using a YOLO model for object detection and saves the output.

    Args:
        input_video_path (str): The path to the input video file.

    Returns:
        str: The path to the output video file.
    """
    # Load the trained model
    model = YOLO("/content/best_activity.pt")

    # OpenCV video reader
    cap = cv2.VideoCapture(input_video_path)

    # Determine output video path
    output_video_path = input_video_path.rsplit('.', 1)[0] + "_processed.mp4"

    # Get video properties
    frame_rate = int(cap.get(cv2.CAP_PROP_FPS))
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

    # Initialize video writer
    fourcc = cv2.VideoWriter_fourcc(*"mp4v")
    out = cv2.VideoWriter(output_video_path, fourcc, frame_rate, (frame_width, frame_height))

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Run inference
        results = model(frame, conf=0.5)

        # Process results and draw bounding boxes
        for r in results:
            for box in r.boxes:
                x1, y1, x2, y2 = box.xyxy[0].cpu().numpy()
                conf = float(box.conf[0])
                cls = int(box.cls[0])
                label = model.names[cls]

                cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
                cv2.putText(frame, f"{label} {conf:.2f}", (int(x1), int(y1)-10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)

        # Write processed frame
        out.write(frame)

    # Release resources
    cap.release()
    out.release()

    return output_video_path
```

## Create gradio interface

### Subtask:
Use the Gradio library to build a user interface around the prediction function, allowing users to upload a video and see the processed output video.


**Reasoning**:
Create a Gradio interface for the `process_video_with_yolo` function.



```python
import gradio as gr

# Create the Gradio interface
iface = gr.Interface(
    fn=process_video_with_yolo,
    inputs=gr.Video(),
    outputs=gr.Video(),
    title="YOLO Activity Detection"
)

# Launch the interface (this would typically be done outside the notebook cell for deployment)
# iface.launch()
```
